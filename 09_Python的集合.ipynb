# 1. Python的集合

① 集合set是一个无序的不重复元素序列。

```python
set1 = set(("小李","小王","小吴")) # 集合是无序的，即输出的顺序不一定和输入时一样
set1
```

② 可以用大括号{}或者set()函数创建集合，但是创建一个空集合必须用set()而不是{}，因为{}是用来创建一个空字典。

```python
list1 = [] # 创建空列表
print(list1)

tuple1 = () # 创建空元组
print(tuple1)

tuple_single = ('1',) # 创建单个元素的元组
print(tuple_single)

dict1 = {} # 创建空字典
print(dict1)

set1 = set() # 创建空集合
print(set1)
print(type(set1))
set1.add(1) # 往集合中添加元素1
print(set1)
print(type(set1))
```

③ 集合增删改查的速度是非常快的，即时间复杂度是很小的。
